```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.info()
```

### The Plot() Method

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots=True,figsize=(15,12),sharex=False,sharey=False)  #sharex= to apply scale to x line
plt.show()                                                             #sharey= to apply scale to y line
                                                                       #-figsize(x,y)--define the boundaries of a plot
```

```python
#ploting age column
titanic.age.plot(figsize=(12,8))
plt.show()
```

### Customization of plots

```python
#Basic requirement to plot the full plot.
titanic.age.plot(figsize=(12,8),fontsize=13, c='r', linestyle="-") #fontsize= fonts size of x ticks and y ticks, c=color  
plt.title("titanic -Ages",fontsize=15)      #linestyle - "-","--",":"    # title -Name of the plot
plt.legend(loc="best",fontsize=15)          # legend - the name for lines. # loc- best/center/right /left
plt.xlabel("Passenger No",fontsize=13)      #xlabel - label for x-axis
plt.ylabel("Age",fontsize=13)               #ylabel- label for y-axis
plt.grid()                                  #Show grid
plt.show()                                  #show plot.
```

```python
#To limit the x and y axis values from 200 to 400 and 20 to 40 #Use xlim=(200,400)  #Use ylim=(20,40)

titanic.age.plot(figsize=(15,5),fontsize=13, c='r', linestyle="-",xlim=(200,400),ylim=(20,40))  
plt.title("titanic -Ages",fontsize=15)    
plt.legend(loc="best",fontsize=15)
plt.xlabel("Passenger No",fontsize=13)
plt.ylabel("Age",fontsize=13)
plt.grid()
plt.show()
```

```python
## to change the x ticks range and y ticks range use xticks and y ticks
# to get the required range ,pass the function shows below and put it in the plot.
titanic.age.plot(figsize=(12,8),fontsize=13, c='r', linestyle="-",xlim=(0,900),ylim=(0,85),xticks=xticks,yticks=yticks) 
plt.title("titanic -Ages",fontsize=15)      
plt.legend(loc="best",fontsize=15)          
plt.xlabel("Passenger No",fontsize=13)      
plt.ylabel("Age",fontsize=13)               
plt.grid()                                  
plt.show() 
```

```python
xticks=[x for x in range(0,901,50)]
xticks
```

```python
yticks=[y for y in range(0,81,5)]
yticks
```

```python
#Rotate the xticks by 45 degrees #USE- rot=45 
titanic.age.plot(figsize=(12,8),fontsize=13, c='r', linestyle="-",xlim=(0,900),ylim=(0,85),xticks=xticks,
                 yticks=yticks,rot=45) 
plt.title("titanic -Ages",fontsize=15)      
plt.legend(loc="best",fontsize=15)          
plt.xlabel("Passenger No",fontsize=13)      
plt.ylabel("Age",fontsize=13)               
plt.grid()                                  
plt.show() 
```

### Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins=10)   #bins= to groups the values in perticular range. Generally to plot histogram
```

```python
titanic.age.plot(kind='hist',figsize=(12,8),fontsize=15,bins=80)
plt.show()
```

```python
titanic.age.hist(figsize=(12,8),bins=80,xlabelsize=15,ylabelsize=15)  # same plot can be used by 'hist' Method
plt.show()
```

```python
# puting age column in plt.hist method  
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(),bins=80,density=False,cumulative=False)  #to put age columnn in hist-we used dropna()
plt.show()                                                    #as hist method do not handle NaN values from age column.
```

```python
# puting age column in plt.hist method  
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(),bins=20,density=True,cumulative=True)
                                                    #density and cumulative is used to plot cum.histogram
plt.show()
```
